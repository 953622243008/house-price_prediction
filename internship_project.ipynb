```python
houses_per_city = df.groupby('city').size().reset_index(name='number_of_houses')

print(houses_per_city)
```

```python
houses_per_city = df.groupby('city').size().reset_index(name='number_of_houses')

houses_per_city_sorted = houses_per_city.sort_values(by='number_of_houses', ascending=False)

print(houses_per_city_sorted)
```

```python
average_price_per_city = df.groupby('city')['price'].mean().reset_index(name='average_price')

average_price_per_city['average_price'] = average_price_per_city['average_price'].astype(int)

average_price_per_city_sorted = average_price_per_city.sort_values(by='average_price', ascending=False)

print(average_price_per_city_sorted)
```

```python
recently_built_houses = df.sort_values(by='yr_built', ascending=False)

print(recently_built_houses)
```

```python
x = df.select_dtypes('number')
plt.figure(figsize=(8,6))
sns.heatmap(x.corr(), annot=True)
```

```python
houses_built_per_year = df['yr_built'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=houses_built_per_year.index, y=houses_built_per_year.values, color='skyblue')
plt.xlabel('Year Built')
plt.ylabel('Number of Houses Built')
plt.title('Number of Houses Built per Year')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
average_price_per_year = df.groupby('yr_built')['price'].mean()

plt.figure(figsize=(10, 6))
average_price_per_year.plot(marker='o', color='skyblue')
plt.xlabel('Year Built')
plt.ylabel('Average House Price')
plt.title('Average House Price Over the Years')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.groupby('city')['price'].mean().sort_values().tail(10).plot(kind="bar")
plt.xlabel("city")
plt.ylabel("avg price")
plt.title("avg prices of housese in each city the highest avg price")
```
